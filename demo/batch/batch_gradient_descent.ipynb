```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from com.sbk.gradientdescent.gradient_descent import batch_gradient_descent
import numpy as np
import pandas as pd
```

```python
init_notebook_mode(connected=True)
```

```python

uploaded_data = pd.read_csv('https://raw.githubusercontent.com/SergiySobolev/gradient_descent/master/data/data.csv')
data = np.asarray(uploaded_data)
x = data[:,0]
y = data[:,1]
theta0, theta1 = batch_gradient_descent(data)
t = np.arange(start=int(min(x))-1, stop= int(max(x))+1, step=0.1)
```

```python
data_trace = go.Scatter(
    x=x,
    y=y,
    mode='markers',
    name='Input data'
)
```

```python
reg_trace = go.Scatter(
    x=t,
    y=theta0 + theta1*t,
    mode='lines',
    name='Linear regression'
)
```

```python
data = [data_trace, reg_trace]

iplot(data)
```
