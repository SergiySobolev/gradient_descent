```python
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
from com.sbk.func.func_1 import calc_gradient_1, func_1
from com.sbk.gradientdescent.gradient_descent import gradient_descent_single_variable


```

# Finding function y = 5(x-3)^4 + 2x - 12 minimum




```python
@interact(start_x=(1.5,4.0,0.1))
def find_and_display_minimum(start_x):
    
    x, f, convergence_path = gradient_descent_single_variable(func_calc_gradient=calc_gradient_1, func=func_1,  start_x=start_x)
    
    plt.rcParams['figure.figsize'] = [15, 7]
    t = np.arange(-1, 7, 0.1)
    plt.plot(t, list(map(func_1, t)))
    plt.plot(start_x, func_1(start_x), "r*", label="Start point")
    plt.plot(x, f, 'o', label="Minimum")
    plt.plot(convergence_path, list(map(func_1, convergence_path)), '+g', label="Convergence path")
    plt.ylim((-7, 3))
    plt.xlim((1.5, 4))
    plt.title("Found minimum", fontweight="bold", color="red")
    plt.legend()
    plt.xlabel("t", fontweight="bold")
    plt.ylabel("y", fontweight="bold")
    plt.show()
```
