```python
import matplotlib.pyplot as plt
import numpy as np
from com.sbk.func.func_1 import calc_gradient, func
from com.sbk.gradientdescent.gradient_descent import gradient_descent_single_variable
```

# Finding function y = 5(x-3)^4 + 2x - 12 minimum




```python
start_x = 1.2
t = np.arange(0, 6, 0.1)
y = func(t)
plt.plot(t, y)
plt.plot(start_x, func(start_x),'rx', label="Start point")
plt.ylim((-15, 50))
plt.title("Initial approximation on y = 5(x-3)^4 + 2x - 12", fontweight="bold", color="red")
plt.xlabel("t", fontweight="bold")
plt.ylabel("y", fontweight="bold")
plt.show()


```

```python
x, f = gradient_descent_single_variable(func_calc_gradient=calc_gradient, func=func,  start_x=start_x)
```

```python
plt.plot(t, y)
plt.plot(start_x, func(start_x),'rx', label="Start point")
plt.plot(x, func(x),'r*', label="Minimum")
plt.ylim((-10, 50))
plt.title("Found minimum", fontweight="bold", color="red")
plt.xlabel("t", fontweight="bold")
plt.ylabel("y", fontweight="bold")
plt.show()



```
