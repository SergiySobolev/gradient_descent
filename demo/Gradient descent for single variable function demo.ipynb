```python
import matplotlib.pyplot as plt
import numpy as np
from com.sbk.gradientdescent.gradient_descent import gradient_descent
```

# Finding function y = 5(x-3)^4 + 2x - 12 minimum



```python
def func(t):
    return 5*(t-3)**4 + 2*t - 12
```

```python
t = np.arange(0, 6, 0.1)
y = func(t)
plt.plot(t, y)
plt.ylim((-15, 50))
plt.title("y = 5(x-3)^4 + 2x - 12", fontweight="bold", color="red")
plt.xlabel("t", fontweight="bold")
plt.ylabel("y", fontweight="bold")
plt.show()

```

#  Gradient grad_y = 4*(x-3)**3 + 2

```python
def calc_gradient(t):
    return 20*(t-3)**3 + 2
```

```python
start_x = 1.2
plt.plot(t, y)
plt.plot(start_x, func(start_x),'rx', label="Start point")
plt.ylim((-15, 50))
plt.title("Initial approximation on plot", fontweight="bold", color="red")
plt.xlabel("t", fontweight="bold")
plt.ylabel("y", fontweight="bold")
plt.show()


```

```python
x, f = gradient_descent(func_calc_gradient=calc_gradient, start_x=start_x)
```

```python
plt.plot(t, y)
plt.plot(start_x, func(start_x),'rx', label="Start point")
plt.plot(x, func(x),'r*', label="Minimum")
plt.ylim((-10, 50))
plt.title("Found minimum", fontweight="bold", color="red")
plt.xlabel("t", fontweight="bold")
plt.ylabel("y", fontweight="bold")
plt.show()

```
