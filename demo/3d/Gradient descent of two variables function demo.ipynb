```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

from com.sbk.func.func_1 import calc_gradient_2, func_2
```

```python
init_notebook_mode(connected=True)
```

```python
t1 = np.arange(-8, 15, 0.5)
t2 = np.arange(-8, 15, 0.5)
```

```python
x,y = np.meshgrid(t1, t2)
z = func_2(x,y)
```

```python
s_a_x = -6
s_a_y = -6
s_a_v = func_2(s_a_x, s_a_y)
```

```python

func_trace = go.Surface(x=x, y=y, z=z, colorscale='Viridis', opacity=0.8)
```

```python
s_a_trace = go.Scatter3d(
    x=[s_a_x],
    y=[s_a_y],
    z=[s_a_v],  
    marker=dict(
        color='red',
        size=5,
        symbol="circle"
    )
)
```

```python
fig = go.Figure(data=[func_trace, s_a_trace])
iplot(fig)
```
