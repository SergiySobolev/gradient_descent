```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

from com.sbk.func.func_1 import calc_gradient_2, func_2, build_table_for_multi_variable_func
```

```python
init_notebook_mode(connected=True)
```

```python
t1 = np.arange(-10, 22, 0.5)
t2 = np.arange(-10, 25, 0.5)
table = np.transpose(build_table_for_multi_variable_func(t1, t2, func_2))

```

```python

func_trace = go.Mesh3d(
                    x=table[0],
                    y=table[1],
                    z=table[2],
                    opacity=0.5,
                    color='rgba(0, 0, 255, 0.3)'
                  )

```

```python
s_a_x = 11
s_a_y = -4
s_a_v = func_2(s_a_x, s_a_y)
```
